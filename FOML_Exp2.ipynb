```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

url = 'https://raw.githubusercontent.com/MohamedZaidhA/Machine-LEarning/refs/heads/main/Salary_dataset.csv'
df = pd.read_csv(url)

df.drop(columns=["Unnamed: 0"], inplace = True)

x="YearsExperience"
y="Salary"

display(df.head())
display(df.shape)
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x=x, y=y, data=df, scatter_kws={'alpha':0.9}, line_kws={'color':'red'})
plt.xlabel('Years of Exp')
plt.ylabel('Salary in Rs.')
plt.grid(True)
plt.show()
```

```python
# Calculate mean
mean_Y = df[y].mean()
mean_X = df[x].mean()
print(f"Mean of y: {mean_Y:.2f}")
print(f"Mean of x: {mean_X:.2f}\n")

# Calculate variance
variance_Y = df[y].var()
variance_X = df[x].var()
print(f"Variance of y: {variance_Y:.2f}")
print(f"Variance of x: {variance_X:.2f}\n")

# Calculate covariance between YearsExperience and Salary
covariance_matrix = df[[x, y]].cov()
covariance = covariance_matrix.loc[x, y]
print(f"Covariance between x and y: {covariance:.2f}")
```

```python
b = covariance / variance_X
a = mean_Y - b * mean_X

print(f"Value of a: {a:.2f}")
print(f"Value of b: {b:.2f}")
```

```python
from sklearn.metrics import mean_absolute_error

# Calculate predicted salaries using the derived 'a' and 'b' values
df['Predicted_Y'] = a + b * df[x]

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(df[y], df['Predicted_Y'])

print(f"Mean Absolute Error (MAE): {mae:.2f}")
```

```python
from sklearn.metrics import mean_squared_error


# Calculate Mean squared Error (MSE)
mse = mean_squared_error(df[y], df['Predicted_Y'])

print(f"Mean Squared Error (MSE): {mse:.2f}")
```

```python
# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```

```python
# Calculate Relative Mean Squared Error
relativeE = mse/variance_Y
print(f"Relative Mean Squared Error (RMSE): {relativeE:.2f}")
```
